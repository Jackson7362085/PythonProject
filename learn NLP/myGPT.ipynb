```python
# start
import re
import os
import pandas as pd

from tqdm import tqdm
import os
def process_text(seq):
    seq=re.sub(r'[_,—,―]','-',seq)
    seq=re.sub(r'\d+[-,.]?\d*','<NUM>',seq)
    return  seq
def process_data(dir='./data/MRPC'):
    data={'train':None,'test':None}
    file_list=os.listdir(dir)
    for f in file_list:
        df=pd.read_csv(os.path.join(dir,f),sep='\t')
        k='train' if 'train' in f else 'test'
        data[k]={'is_same':df.iloc[:,0].values,'s1':df.iloc[:,3].values,
                 's2':df.iloc[:,4].values}
    vocab=set()
    for k in ['train','test']:
       for m in ['s1','s2']:
            for i,item in tqdm(enumerate(data[k][m])):
                text=process_text(item.lower())
                data[k][m][i]=text
                cs=text.split(" ")
                vocab.update(set(cs))
    v2i={v:i for i,v in enumerate(vocab,start=1)}

    v2i['<MASK>']=len(v2i)
    v2i['<GO>']=len(v2i)
    v2i['<SEP>']=len(v2i)
    i2v={i:v for v,i in v2i.items()}
    for k in ['train','test']:
        for m in ['s1','s2']:
            data[k][m+'id']=[[v2i[c] for c in seq.split(" ")]for seq in data[k][m]]

    return data,i2v,v2i
```

```python
PAD_ID=0
import numpy as np
def pad_zero(seqs, max_len):
    padded = np.full((len(seqs), max_len), fill_value=PAD_ID, dtype=np.int32)
    for i, seq in tqdm(enumerate(seqs)):
        print(seq)
        padded[i, :len(seq)] = seq
    return padded
```

```python
class MRPCData:
    def __init__(self,dir='./data/MRPC'):
        data,self.i2v,self.v2i=process_data(dir)
        self.x_len=[[len(data['train']['s1id'][i]),len(data['train']['s2id'][i])] for i in range(len(data['train']['s1id']))]
        x=[[self.v2i['<GO>']]+data['train']['s1id'][i]+[self.v2i['<SEP>']]
            +data['train']['s2id'][i]+[self.v2i['<SEP>']]for i in range(len(self.x_len))]
        max_len=max([len(s1)+len(s2)+3 for s1,s2 in zip(data['train']['s1id']+data['test']['s1id'],data['train']['s2id']+data['test']['s2id'])])
        self.x = pad_zero(x, max_len=max_len)
        self.flag=data['train']['is_same'][:,None]
        self.judge=np.full(self.x.shape,2,dtype=np.int32)
        for i in range(len(self.x)):
            si=self.x_len[i][0]+2
            self.judge[i][:si]=0
            _si=self.x_len[i][1]+1
            self.judge[i][si:_si]=1
    def sample(self,n):
        xi=np.random.randint(0,self.x.shape[0],n)
        a,b,c,d=self.x[xi],self.judge[xi],self.x_len[xi],self.flag[xi]





```
